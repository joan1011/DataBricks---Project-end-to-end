# Init

```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

# Business Tranformation and Modeling

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY ci.customer_id) AS customer_key,
    ci.customer_id,
    ci.firstname,
    ci.lastname
FROM
  silver.crm_customers AS ci
"""
df = spark.sql(query)   
```

# Write it to Gold Table

```python
(
    df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("gold.dim_customers")
)
```
